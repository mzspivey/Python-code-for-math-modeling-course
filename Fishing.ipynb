```python
# This is an implementation of a simple model of a fish population with a constant growth rate and a harvesting term.

import matplotlib.pyplot as plt

rate = 0.3
imax = 50
harvested = 290  # Number of fish removed each year
fish = [0]*imax

fish[0] = 1000
for i in range(imax-1):
    fish[i+1] = (1+rate)*fish[i] - harvested   # Difference equation model with harvesting

plt.plot(fish)
fish[imax-1]
```
