```python
# How do you compare Olympic lifters across different weight classes?
# A second hypothesis, based on a geometric similarity argument, is that max lift should be proportional to the max weight to the 2/3 power.
# This code visually tests that hypothesis as a model.

import matplotlib.pyplot as plt

WeightMen = [56, 62, 69, 77, 85, 94, 105]
LiftMen = [305, 325, 357.5, 376.5, 390, 405, 425]

WeightWomen = [48, 53, 63, 75]
LiftWomen = [185, 225, 242.5, 245]

# I removed the largest category because it didn't have a true max weight.

plt.scatter(WeightMen,LiftMen)  # These produce scatterplots.
plt.scatter(WeightWomen,LiftWomen)

plt.figure() 
MenWTwoThirds = [W**(2/3) for W in WeightMen]
WomenWTwoThirds = [W**(2/3) for W in WeightWomen]
plt.scatter(MenWTwoThirds, LiftMen)
plt.scatter(WomenWTwoThirds, LiftWomen)
```

```python
plt.figure()
HandicapMen = [L / W for L, W in zip(LiftMen, MenWTwoThirds)]
HandicapWomen = [L / W for L, W in zip(LiftWomen, WomenWTwoThirds)]
plt.scatter(WeightMen, HandicapMen)
plt.scatter(WeightWomen, HandicapWomen)
```
