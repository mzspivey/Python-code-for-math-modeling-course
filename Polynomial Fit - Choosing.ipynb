```python
# How do you decide which degree polynomial is the best fit for a data set?
# This code illustrates the use of training and validation sets to help address that question.

import numpy as np
import matplotlib.pyplot as plt

xtrain = [17, 19, 20, 22, 23, 25, 28, 31, 32, 33, 36, 37, 39, 42]
ytrain = [19, 25, 32, 51, 57, 71, 113, 140, 153, 187, 192, 205, 250, 260]
          
xvalid = [17, 17, 19, 19, 25, 25, 28, 31, 32, 38, 39, 39, 41]
yvalid = [16, 22, 28, 21, 56, 86, 88, 141, 123, 252, 231, 265, 294]

plt.scatter(xtrain,ytrain)
```

```python
poly1 = np.polynomial.polynomial.polyfit(xtrain,ytrain,1) 
predict1 = np.polynomial.polynomial.polyval(xvalid,poly1)
resid1 = [y - x for y,x in zip(yvalid,predict1)]
plt.plot(xvalid,predict1)
plt.scatter(xvalid,yvalid)
poly1
```

```python
def sumsquares(lst):
    lst_sq = [x**2 for x in lst]
    return sum(lst_sq)
```

```python
ssq = []
for i in range(0,17):  
    poly = np.polynomial.polynomial.polyfit(xtrain,ytrain,i) 
    predict = np.polynomial.polynomial.polyval(xvalid,poly)
    resid = [y - x for y,x in zip(yvalid,predict)]
    ssq.append(sumsquares(resid))
ssq
```

```python
plt.plot(ssq[1:10])  # Let's rule out a constant model, as well as the really high-order models.
```

```python
xplot = np.linspace(17,42,100) 
poly2 = np.polynomial.polynomial.polyfit(xtrain,ytrain,2) 
predict2 = np.polynomial.polynomial.polyval(xplot,poly2)
plt.plot(xplot,predict2)
plt.scatter(xvalid,yvalid)
```
