```python
# This is a logistic model of a fishing population featuring a constrant growth rate, a carrying capacity, and a harvesting term.

import matplotlib.pyplot as plt

rate = 0.3
imax = 50
harvested = 180
limit = 2500
fish = [0]*imax

fish[0] = 1300
for i in range(imax-1):
    fish[i+1] = fish[i] + rate*fish[i]*(1-fish[i]/limit) - harvested

plt.plot(fish)
fish[imax-1]
```
