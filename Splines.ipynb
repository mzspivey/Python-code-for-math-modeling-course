```python
# This code shows how to fit linear and cubic splines to a data set.

import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

speed = [x for x in range(0,17)]
StoppingData = [0, 3, 6, 11, 21, 32, 47, 65, 87, 112, 140, 171, 204, 241, 282, 325, 376]

plt.plot(speed,StoppingData)
plt.scatter(speed,StoppingData)
```

```python
f1 = interpolate.interp1d(speed,StoppingData)
f1([4.5, 8.9, 11])
```

```python
f3 = interpolate.interp1d(speed,StoppingData,"cubic")
plotspeed = np.linspace(0,16,100)
plt.plot(plotspeed,f3(plotspeed))
plt.scatter(speed,StoppingData)
```

```python
diameter = [17, 19, 20, 22, 23, 25, 28, 31, 32, 33, 36, 37, 39, 42]
boardfeet = [19, 25, 32, 51, 57, 71, 113, 140, 153, 187, 192, 205, 250, 260]
```

```python
p1 = interpolate.interp1d(diameter, boardfeet)
p3 = interpolate.interp1d(diameter, boardfeet, "cubic")
plotdiameter = np.linspace(17,42,100)
plt.plot(plotdiameter, p1(plotdiameter))
plt.plot(plotdiameter, p3(plotdiameter))
plt.scatter(diameter,boardfeet)
```
