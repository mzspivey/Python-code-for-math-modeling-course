```python
# This is an example of a system of difference equations - a predator-prey model.

import matplotlib.pyplot as plt

max = 100
rabbit = [0]*max
fox = [0]*max

rabbit[0] = 100
fox[0] = 100

A = 0.2
B = 0.001
C = 0.2
D = 0.002

rabbit[0] = C/D + 1
fox[0] = A/B + 1

for i in range(max-1):
    rabbit[i+1] = (1+A)*rabbit[i] - B*rabbit[i]*fox[i]
    fox[i+1] = (1-C)*fox[i] + D*rabbit[i]*fox[i]
    
plt.plot(rabbit)
plt.plot(fox)
```
