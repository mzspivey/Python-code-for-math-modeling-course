```python
# This is a simple example of a discrete dynamical system: a compound interest model.

import matplotlib.pyplot as plt

rate = -1.5
imax = 50
amount = [0]*50   # Initialize list of size 50 and set all values to 0.  The indexes range from 0 to 49.

amount[0] = -.10   # Initialize first list value to 1000

for i in range(imax-1):   # Iterate from i = 0 to i = imax-2.
    amount[i+1] = (1+rate)*amount[i]  # Difference equation

plt.plot(amount)
```
