```python
# This code illustrates the use of data transforms for fitting a data set to different models.

import matplotlib.pyplot as plt
import math

speed = [x for x in range(0,17)]
StoppingData = [0, 3, 6, 11, 21, 32, 47, 65, 87, 112, 140, 171, 204, 241, 282, 325, 376]

plt.scatter(speed, StoppingData)
```

```python
# Test for exponential model
logspeed = [math.log(x) for x in speed[1:]]  # Why are we doing this?  This does the transformation on all values but the first in the list, in order to avoid log(0).
plt.scatter(logspeed, StoppingData[1:])
```

```python
# Test for a logarithmic model
logstop = [math.log(x) for x in StoppingData[1:]]
plt.scatter(speed[1:], logstop)
```

```python
#Test for power model
plt.scatter(logspeed, logstop)
```
